# **NumPy(Array Operations)**
# **1.**

```python
import numpy as np
arr=np.random.randint(1,51,size=(5,5)) # Creating array with random integers from 1 to 50
rsum=np.sum(arr,axis=1) # Creating an array with all the row sums
rindex=np.argmax(rsum) # Getting the index of the largest element
print("Original Array:")
print(arr)
print("Row with maximum sum:",arr[rindex]) # Displaying the row with largest sum
mean=np.mean(arr) # Finding the mean
arr[arr>mean]=0 # Creating an array with boolean values and replacing True with 0
print("Modified Array:")
print(arr)
```

# **2.**

```python
arr = np.arange(1, 21) # Creating an array with elements from 1 to 20
print("Original Array:", arr)
new_arr = arr[1::3] # 2nd element=element with index 1 and getting all 3rd elements starting from index 1
print("New Array:", new_arr)
```

# **Pandas(Data Handling)**
# **3.**

```python
import pandas as pd
import seaborn as sns
df=sns.load_dataset("iris") # loading the dataset using seaborn
print("First 10 rows:")
print(df.head(10)) # Getting the 1st 10 rows using head()
print("\n")
print("Basic Statistics:") # Getting the basic statistics using describe()
print(df.describe())
print("\n")

sepal_median = df['sepal_length'].median()
filtered_df = df[(df['sepal_length'] > sepal_median) & (df['species'] == 'versicolor')] # Accessing the filtered dataframe
print("Filtered Rows:")
print(filtered_df)
print("\n")


grouped_mean=df.groupby('species')['petal_length'].mean().reset_index() # Grouping based on species and getting all the mean of petal length
print("Grouped Mean:")
print(grouped_mean)
```

# **4.**

```python
df['petal_ratio']=df['petal_length']/df['petal_width'] # Creating new column named petal_ratio
top5=df.sort_values(by='petal_ratio',ascending=False).head(5) # Sorting and getting the top 5 with the highest petal_ratio
print("Top 5 Rows with highest petal_ratio:")
print(top5)
```

# **Data Visualization**
# **5.**

```python
import matplotlib.pyplot as plt # importing matplotlib
plt.figure(figsize=(8, 5))
plt.hist(df['sepal_length'], bins=15, color='yellow', edgecolor='black') # Customising the histogram
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()


```

```python
colors = {'setosa': 'red', 'versicolor': 'blue', 'virginica': 'green'} # Giving colours to each species through a dictionary
plt.figure(figsize=(8, 6))
for species, group in df.groupby('species'): # Using for loop for setting labels for each species to use legends later
    plt.scatter(
        group['sepal_length'],
        group['petal_length'],
        c=colors[species],
        label=species,
        alpha=0.8
    )
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend(title='Species')
plt.show()
```
