# **1.Simple ATM Calculator**

```python
Balance=int(input("Balance = ")) #Getting Balance from the user
Withdrawal=int(input("Withdrawal = ")) #Getting withdrawal amount from the user
# Checking if the withdrawal amount exceeds the limit
if Balance>=Withdrawal:
  print("Transaction Successful, Remaining Balance = ₹",Balance - Withdrawal)
else:
  print("Error!, Withdrawal exceeds the Balance")
```

# **2.Electricity Bill Generator**

```python
Units=int(input("Units Consumed= ")) # Getting units comsumed from the user
if Units<=100: # For the 1st 100 units, it is Rs.5 per unit
  print("Bill amount = ₹",Units*5,"(",Units, "x 5)")
elif Units<=200: # For units b/w 100 & 200, multiply 100 with 5 and the remaining value with 7
  print("Bill amount = ₹",100*5+(Units-100)*7, "(",100,"x 5 +",Units-100,"x 7)")
else: # For units greater than 200, multiply the 1st 100 units with 5, next 100 units with 7 and the remaining value with 10
  print("Bill amount = ₹",100*5+100*7+(Units-200)*10,"(",100,"x 5 +",100,"x 7 +",Units-200,"x 10)")
```

# **3.Temperature Converter**

```python
Temp=int(input("Temperature =")) # Getting the temperature from the user
Unit=(input("Unit =")) # Getting the unit from the user
if Unit=="C" or Unit=="c": # To make it into case insensitive
  print("Temperature in Fahrenheit = ",(Temp*9/5)+32,"°F") # Converting the temperature to fahrenheit with the formula
elif Unit=="F" or Unit=="f": # To make it into case insensitive
  print("Temperature in Celsius = ",(Temp-32)*5/9,"°C") # Converrting the temperature to celsius with the formula
```

# **4.Mobile Recharge Validity**

```python
from datetime import datetime # importing datetime package to get the required date functions
Re_date=datetime.strptime(input("Recharge date (YYYY-MM-DD): "), "%Y-%m-%d") # Using strptime() to access the date from the user
Today=datetime.strptime(input("Today's date (YYYY-MM-DD): "), "%Y-%m-%d")
if (Today-Re_date).days<=28: # Checking whether the dates have a difference of 28 days or less
  print("Plan is still available")
else:
  print("Plan expired. Recharge again")
```

# **6.Sum of Even Numbers**

```python
Num=int(input("Number=")) # Getting the range number from the user
Sum=0
for i in range(1,Num+1): # using for loop for an iterative process
  if i%2==0: # Checking whether if the given number is even or not by dividing it by 2 and checking the remainder
    Sum=Sum+i # When the condition satisfies add the number to the sum
print("Sum of even numbers upto",Num," = ",Sum)
```

# **8.Dictionary-Word Frequency**

```python
sentence=input("Enter a sentence:").lower().split() # getting a sentence from the user, splitting into words and converting into lower case
Word_dic={} # Creating the dictionary to store the words and counts
for word in sentence:
  Word_dic[word]=Word_dic.get(word,0)+1 # Checking the excistence of the given word using get function
for key,value in Word_dic.items(): # Printing the dictionary items
  print(key,":",value)
```

# **9.Functions-Check Prime**

```python
import math # importing math module to get the sqrt function
def prime(num):
  if num<=1: # number less than or equal to 1 is not a prime number
    return False
  if num==2: # the only even prime number
    return True
  if num%2==0: # even numbers cannot be prime numbers except 2
    return False
  sqrt=int(math.sqrt(num)) # getting the square root of the given number
  for i in range(3,sqrt + 1,2): # if there is no divisor upto the square root of the number, then it is a prime
    if num%i==0:
      return False # found a divisor, not a prime number
  return True # no divisor, prime number
number=int(input("Enter a number:"))
Check=prime(number)
if Check==True:
  print(number," is a prime number")
else:
  print(number," is not a prime number")
```

# **10.Vowel Counter**

```python
str=input("Enter a string:")
count=0
for i in range(0,len(str)):
  if str[i].lower() in 'aeiou':
    count+=1
print("Number of Vowels:",count)
```

# **11.Count Divisible Numbers**

```python
limit=int(input("Enter the range limit:"))
div=int(input("Enter the divisor:"))
count=0
for i in range(1,limit+1):
  if i%div==0:
    count+=1
print("Count of numbers divisible by ",div," : ",count)
```
